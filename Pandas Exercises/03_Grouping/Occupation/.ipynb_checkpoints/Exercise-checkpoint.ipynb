### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

users = pd.read_csv(address, delimiter='|')
```

### Step 4. Discover what is the mean age per occupation

```python
round(users.groupby('occupation')['age'].mean(),2)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#function to convert gender to number so we can identify males more easily
def gender_to_number(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

#applying thefuntion to a new columns
users['gender_n'] = users['gender'].apply(gender_to_number)

#maths to get the ratio
a = round(users.groupby('occupation')['gender_n'].sum() / users['occupation'].value_counts(),2) * 100

#sorting and printing
a.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].agg(['mean'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
c = pd.DataFrame()

#maths to get the ratio
c['Males %'] = a

c['Females %'] = 100 - c['Males %']
#sorting and printing
c
```
